### 📌 Creating a Local Temporary View
Local temporary views are visible only to the session that created them and are dropped when the session ends.

```python
df = spark.read.table("population_metrics.default.countries_population")
```

```python
df.createOrReplaceTempView("_sql_df")
```

```python
spark.sql("select * from _sql_df").display()
```

### 📌 Creating a Global Temporary View
To create a global temporary view you use the createGlobalTempView or createOrReplaceGlobalTempView method on the dataframe and give the view a name.


```python
df.createOrReplaceGlobalTempView("_sql_gdf")
```

```python
spark.sql("select * from global_temp._sql_gdf")
```

```python
# run in other notebook
spark.sql("select * from global_temp._sql_gdf")
```
