### 📌 Syntax

```python
%sql

-- Dropping a catalog

DROP CATALOG [ IF EXISTS ] catalog_name [ RESTRICT | CASCADE ]
```

```python
%sql
DROP CATALOG test_catalog CASCADE
```

```python
%sql

-- Dropping a schema

DROP SCHEMA [ IF EXISTS ] schema_name [ RESTRICT | CASCADE ]
```

```python
%sql
DROP SCHEMA test_catalog.test_schema CASCADE
```

```python
%sql

-- Dropping a Volume

DROP VOLUME [ IF EXISTS ] volume_name
```

```python
%sql

-- Dropping a Table

DROP TABLE [ IF EXISTS ] table_name
```

```python
%sql

-- Dropping a View

DROP VIEW [ IF EXISTS ] view_name
```
