### 📌 Creating a View
`CREATE VIEW <catalog-name>.<schema-name>.<view-name> AS
SELECT <query>;`

```python
%sql

-- Creating a demo view

CREATE VIEW population_metrics.default.top_10_countries_population AS
SELECT * FROM population_metrics.default.countries_population ORDER BY population DESC LIMIT 10;
```

### 📌 Querying a View

```python
%sql

-- Querying the view using SQL

SELECT * FROM population_metrics.default.top_10_countries_population
```

### 📌 Returning the view as a DataFrame

```python
# Returning the view as a DataFrame object via the spark.sql() method

spark.sql("SELECT * FROM population_metrics.default.top_10_countries_population").display()
```

```python
# Returning the view as a DataFrame object via the table() method

spark.read.table("population_metrics.default.top_10_countries_population").display()
```
