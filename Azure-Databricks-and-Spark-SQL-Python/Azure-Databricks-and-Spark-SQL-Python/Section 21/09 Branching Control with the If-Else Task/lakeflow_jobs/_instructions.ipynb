### 📌 Instructions

Create a job called **if_else_demo**.

In this directory you’ll find three notebooks scripts:

1. **`run_if_true`**  
2. **`run_if_false`**  
3. **`branching_control_notebook`**

Create a task called **if_else_task** and experiment with various conditions that evaluate to either true or false.

- Assign the `True` branch to run a Notebook Task that executes **run_if_true**.
- Assign the `False` branch to run a Notebook Task that executes **run_if_false**.


You can also provide conditions via job parameters or even define a `taskValue` in an upstream Notebook Task:
- Use the **branching_control_notebook** to reference the `taskValue` in the If-else condition task.
