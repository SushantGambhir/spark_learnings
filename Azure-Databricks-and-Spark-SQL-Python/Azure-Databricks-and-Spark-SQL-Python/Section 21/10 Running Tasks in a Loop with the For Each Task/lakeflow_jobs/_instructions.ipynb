### 📌 Instructions

Create a job called **for_each_demo**.

In this directory you’ll find two notebooks scripts:

1. **`iterative_notebook`**  
2. **`input_notebook`**  

Create a task called **for_each_task** and provide a list of regions as input values.

The nested task inside of the **for_each_task** should execute the **iterative_notebook**.

You can also provide the input list via job parameters or even define a `taskValue` in an upstream Notebook Task:
- Use the **input_notebook** to reference the `taskValue` in the input for the for_each task.
