### 📌 Reading data into a Spark DataFrame

```python
df = spark.read.table("population_metrics.default.countries_population")
```

### 📌 Examples

```python
from pyspark.sql.functions import length, char_length, initcap, upper, lower, concat, concat_ws, lit
```

```python
#length()

display(df.select(
    "name",
    length("name").alias("count_of_chars")
))
```

```python
# char_length()

display(df.select(
    "name",
    char_length("name").alias("count_of_chars")
))
```

```python
# char_length()

display(df.select(
    "name",
    char_length("name").alias("count_of_chars")
))
```

```python
# upper() and lower()

display(df.select(
    "name",
    upper("name").alias("name_upper"),
    lower("name").alias("name_lower")
))
```

```python
#lit()

display(df.select(
    "name",
    lit("hello")
))
```

```python
# concat() and concat_ws()

display(df.select(
    "name",
    "capital",
    concat("capital", lit(", "), "name").alias("city_and_country"), # here we need to specify the separator as literal at position(s) we want
    concat_ws(", ", "name", "capital").alias("city_and_country_ws") # here we have to specify separator as first argument directly as a string
    ))
```
