### 📌 Creating a Spark DataFrame

```python
from pyspark.sql.functions import to_date, to_timestamp, date_format
```

```python
schema = "name string, hire_date string"

# Create a minimal DataFrame with name and hire_date
data = [
    ("Alice", "15-06-2020"),
    ("Bob", "25-09-2018"),
    ("Charlie", "05-12-2022")
    ]

df = spark.createDataFrame(data, schema)

df = df.withColumn("hire_date", to_date("hire_date", "dd-MM-yyyy"))

display(df)
```

### 📌 Examples

```python
from pyspark.sql.functions import curdate, current_timestamp, timestamp_diff
```

```python
# curdate() and current_timestamp()
 
df = df.withColumn("last_updated_date", curdate()).withColumn("last_updated_timestamp", current_timestamp())

display(df)
```

```python
# timestamp_diff

display(df.withColumn("days_since_hired", timestamp_diff("DAY", "hire_date", "last_updated_date")))
```

```python
display(df.withColumn("days_since_hired", timestamp_diff("MONTH", "hire_date", "last_updated_date")))
```
