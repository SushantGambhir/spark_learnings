### 📌 Standard Date Format Patterns

Date: `yyyy-MM-dd`

Timestamp: `'yyyy-MM-dd[T]HH:mm:ss[.SSSSSS]`


### 📌 Creating a Spark DataFrame

```python
schema = "name string, hire_date string"

# Create a minimal DataFrame with name and hire_date
data = [
    ("Alice", "15-06-2020"),
    ("Bob", "25-09-2018"),
    ("Charlie", "05-12-2022")
    ]

df = spark.createDataFrame(data, schema)

df.printSchema

display(df)
```

### 📌 Examples

```python
from pyspark.sql.functions import to_date, to_timestamp, date_format
```

```python
df = df.withColumn('hire_date_converted',to_date('hire_date','dd-MM-yyyy'))
```

```python
df.printSchema()
display(df)
```

```python
# to_timestamp()

df = df.withColumn("hire_timestamp_converted", to_timestamp("hire_date", "dd-MM-yyyy"))

display(df)
```

```python
# date_format()

display(df.withColumn("year", date_format("hire_date_converted", "yyyy-MM"))) 
```

```python
display(df.withColumn("month_year", date_format("hire_date_converted", "MMMM, yyyy"))) 
```
