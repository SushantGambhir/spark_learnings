```python
%sql
-- Creating an empty Managed Table
CREATE TABLE population_metrics.default.countries_population_2
(
  country_id integer, 
  name string, 
  nationality string, 
  country_code string, 
  iso_alpha2 string, 
  capital string, 
  population integer, 
  area_km2 integer, 
  region_id integer, 
  sub_region_id integer
 )
```

```python
# Returning the newly created managed table as a DataFrame

df = spark.sql("select * from population_metrics.default.countries_population_2")

df.display()
```

```python
# Inserting records into the table from the countries_population table

df = spark.sql("select * from population_metrics.default.countries_population")

df.write.mode("append").saveAsTable("population_metrics.default.countries_population_2")
```

```python
spark.sql("select * from population_metrics.default.countries_population_2").display()
```

```python
%sql

-- CREATE TABLE AS Statement can create a table from a query result

CREATE TABLE population_metrics.default.countries_population_3
AS SELECT * FROM population_metrics.default.countries_population_2
```
