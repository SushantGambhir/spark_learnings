### 📌 Reading data into a Spark DataFrame

```python
df = spark.read.table("population_metrics.default.countries_consolidated")

display(df)
```

### 📌 Examples

```python
# Using column instances to filter the DataFrame
display(df.filter(df.region == "Asia"))
```

```python
# This uses SQL expression-based filtering.
display(df.filter("region = 'Asia'"))
```

```python
# filtering on multiple conditions using the | operator
display(df.filter((df.region == "Asia") | (df.region == "Europe")))
```

```python
# filtering on multiple conditions using the 'or' operator
display(df.filter("region = 'Asia' or region = 'Europe'"))
```

```python
# filtering on multiple conditions using the & operator
display(df.filter((df.region == "Asia") & (df.country == "India")))
```

```python
# filtering on multiple conditions using the and operator
display(df.filter("region = 'Asia' and country = 'India'"))
```

```python
# Using an expression to filter records (column instances)
display(df.where((df.population/df.area_km2) > 1000))
```

```python
# Using an expression to filter records (SQL string)
display(df.where("(population/area_km2) > 1000"))
```
