### 📌 Reading data into a Spark DataFrame

```python
df = spark.read.table("population_metrics.default.countries_consolidated")

display(df)
```

### 📌 Examples

```python
# sort on population (ascending by default)

display(df.sort("population"))
```

```python
# sort on population descending

display(df.sort("population", ascending=False))
```

```python
# sort on population and limit to top 10 records

display(df.sort("population", ascending=False).limit(10))
```

```python
# sort on region and population (both descending)

display(df.sort("region", "population", ascending=False))
```

```python
# sort on regsion ascending and population descending 
# In this way we must necessarily give column objects, string values will not work unlike above

display(df.sort(df.region.asc(), df.population.desc()))
```

```python
# Alternatively, the above statement can also be used as

display(df.sort("region", "population", ascending=[True,False]))
```

```python
# orderBy is an alias for sort

display(df.orderBy(df.region.asc(), df.population.desc()))
```
